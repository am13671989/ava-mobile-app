## Step 2: Run Trainer Script

This cell finds the uploaded trainer script automatically, even if Colab renamed it with `(1)`.

````python
import glob

scripts = sorted(glob.glob('USE_THIS_TFLITE_TRAINER_IN_COLAB*.py'))
print('Found scripts:', scripts)

if not scripts:
    raise FileNotFoundError('Upload USE_THIS_TFLITE_TRAINER_IN_COLAB.py first.')

script_path = scripts[-1]
print('Running:', script_path)
exec(open(script_path).read())
````
